```python
import requests
from bs4 import BeautifulSoup

l_ulr2p = [ 'https://pinturachilena.cl/explora/artistas',
        ]
headers = {"User-Agent": "Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:108.0) Gecko/20100101 Firefox/108.0"}

response = requests.get(url=l_ulr2p[0], headers=headers)
soup2p = BeautifulSoup(response.text, "lxml")
overview = soup2p.find()
print(overview)
```

```python
## Ver HTML  

ejemplo2p = soup2p.find_all("h4", class_="case27-secondary-text") 
# mostrar
#print( len(ejemplo2p) )
```

```python
# limpiar lista

dela = "<h4 class=\"case27-secondary-text\">"
delb = "</h4>"

l_pintores = []

for i in range( len(ejemplo2p) ): 
    #print( type(ejemplo2p[i].text) )  
    nomb = ejemplo2p[i].text 
    l_pintores.append( nomb )

l_pintores
```

```python
# Guardar datos pintores
# diccionario de listas de los contenidos
dict_pintorescl_nombres_01 = { 'pintores_nombres': l_pintores } 

# Diccionario a dataframe 
dfpintorescl_nombres_01 = pd.DataFrame(dict_pintorescl_nombres_01)

# Guardar a excel
dfpintorescl_nombres_01.to_excel("a_auras_pintorescl_nombres_01.xlsx.xlsx")


```
